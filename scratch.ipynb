```python
import numpy as np
```

```python
input_image = np.load("./data/image_65.npy")
dense_weights = np.load("./data/dense_weights.npy")
dense_biases = np.load("./data/dense_biases.npy")
dense_biases = dense_biases.reshape(-1,1)

dense_1_weights = np.load("./data/dense_1_weights.npy")
dense_1_biases = np.load("./data/dense_1_biases.npy")
dense_1_biases = dense_1_biases.reshape(-1,1)
```

```python
dense_weights.shape
```

```python
input_image.shape
```

```python
input_image
```

```python
dense_biases.shape
```

```python
l1 = np.matmul(dense_weights.transpose(), input_image)
```

```python
l1
```

```python
b1 = (l1+dense_biases)
b1
```

```python
b1_relu = (np.abs(b1)+b1)/2
b1_relu
```

```python
l2 = np.matmul(dense_1_weights.transpose(), b1_relu)
l2
```

```python
b2 = (l2+dense_1_biases)
b2
```

```python
import tensorflow as tf
b2_ = tf.nn.softmax(b2.squeeze())
b2_
```
